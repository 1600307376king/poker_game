```python
import matplotlib.pyplot as plt
import numpy as np


def show_normalize(x, y):
    plt.plot(x, y)
    plt.show()

def toGas(x):
    u1 = 0
    sigmal = 1
    # a = 0.1
    # b = 0
    # c = 5
    return np.multiply(np.power(np.sqrt(2*np.pi)*sigmal, -1), np.exp(-np.power(x - u1, 2) / 2 * sigmal ** 2))*20

def toCos(x):
    return 10* np.cos(x) + 10

if __name__ == '__main__':
    x_arr = np.arange(-1, 4, 1).astype(np.float32)
    show_normalize(x_arr, toGas(x_arr))
```
